# **Image masking for computer vision using python**

> In computer vision, masks are binary images that highlight specific regions or objects within an image. These masks are essentially a way of separating and identifying the areas of interest.
Objectives:
*   Object tracking
*   object segmentation

â˜£ Image masking is to combine original image with the segmented labels obtained using tools (ultrack software and other tools or manually) .

```python
# import the requirements
import cv2
#if running locally on vscode or jupyter notebook then do not use this line and simply replace the cv2_imshow() in second cell with cv2.imshow()
from google.colab.patches import cv2_imshow
```

```python
#load the original image
original_image = cv2.imread('/content/drive/MyDrive/t000.jpg')
cv2_imshow(original_image)
```

```python
# load the segmented labelled image
segmentation_mask = cv2.imread('/content/drive/MyDrive/t000.png')
cv2_imshow(segmentation_mask)
```

```python
# Ensure that the segmentation mask has the same dimensions as the original image
segmentation_mask = cv2.resize(segmentation_mask, (original_image.shape[1], original_image.shape[0]))
```

```python
# Apply the segmentation mask to the original image
result = cv2.bitwise_and(original_image, segmentation_mask)
```

```python
# Display the result
cv2_imshow(result)
cv2.waitKey(0)
```

```python
# Save the result
cv2.imwrite('overlayed_image.jpg', result)

# Close all OpenCV windows
cv2.destroyAllWindows()
```
